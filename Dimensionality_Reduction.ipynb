```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
from sklearn.manifold import TSNE
```

First we load the data

```python
data = np.load('data.npy')

print(data.shape)

labels = np.load('labels.npy')
print(labels.shape)
```

Then we are applying dimensional reduction on it to go from 6 dimensions to 2 and 3 dimensions using pca,mds and t-sne

```python
# PCA 2D
pca_2d = PCA(n_components=2)
data_pca_2d = pca_2d.fit_transform(data)
print(data_pca_2d.shape)

# PCA 3D
pca_3d = PCA(n_components=3)
data_pca_3d = pca_3d.fit_transform(data)
print(data_pca_3d.shape)

# MDS 2D
mds_2d = MDS(n_components=2, normalized_stress='auto')
data_mds_2d = mds_2d.fit_transform(data)
print(data_mds_2d.shape)

# MDS 3D
mds_3d = MDS(n_components=3, normalized_stress='auto')
data_mds_3d = mds_3d.fit_transform(data)
print(data_mds_3d.shape)


# t-SNE 2D
tsne_2d = TSNE(n_components=2, random_state=42)
data_tsne_2d = tsne_2d.fit_transform(data)
print(data_tsne_2d.shape)

# t-SNE 3D
tsne_3d = TSNE(n_components=3, random_state=42)
data_tsne_3d = tsne_3d.fit_transform(data)
print(data_tsne_3d.shape)
```

And finally we are displaying each the results of each method in 2d and 3d one by one

```python
# PCA 3D

sns.scatterplot(x=data_pca_2d[:, 0], y=data_pca_2d[:, 1], hue=labels, palette='viridis', edgecolor='k')
plt.title('PCA 2 components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# MDS 3D

sns.scatterplot(x=data_mds_2d[:, 0], y=data_mds_2d[:, 1], hue=labels, palette='viridis', edgecolor='k')
plt.title('MDS 2 components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# T-SNE 2D

sns.scatterplot(x=data_tsne_2d[:, 0], y=data_tsne_2d[:, 1], hue=labels, palette='viridis', edgecolor='k')
plt.title('t-SNE 2 components')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
# PCA 3D
fig = go.Figure(data=[go.Scatter3d(
    x=data_pca_3d[:, 0], y=data_pca_3d[:, 1], z=data_pca_3d[:, 2],
    mode='markers',
    marker=dict(
        size=2,
        color=labels,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)])

fig.update_layout(title='PCA 3 components', scene=dict(
    xaxis_title='Principal Component 1',
    yaxis_title='Principal Component 2',
    zaxis_title='Principal Component 3'))
fig.show()
```

```python
# MDS 3D
fig = go.Figure(data=[go.Scatter3d(
    x=data_mds_3d[:, 0], y=data_mds_3d[:, 1], z=data_mds_3d[:, 2],
    mode='markers',
    marker=dict(
        size=2,
        color=labels,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)])

fig.update_layout(title='MDS 3 components', scene=dict(
    xaxis_title='Principal Component 1',
    yaxis_title='Principal Component 2',
    zaxis_title='Principal Component 3'))
fig.show()
```

```python
# T-SNE 3D

fig = go.Figure(data=[go.Scatter3d(
    x=data_tsne_3d[:, 0], y=data_tsne_3d[:, 1], z=data_tsne_3d[:, 2],
    mode='markers',
    marker=dict(
        size=2,
        color=labels,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)])

fig.update_layout(title='t-SNE 3 components', scene=dict(
    xaxis_title='Component 1',
    yaxis_title='Component 2',
    zaxis_title='Component 3'))
fig.show()
```

As we can see, for this dataset the method that seems to suit the best is indeed the PCA
